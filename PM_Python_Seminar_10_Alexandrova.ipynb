```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
# Задача 63.
# 1. Изобразите отношение households к population с
# помощью точечного графика

sns.scatterplot(data=df, x='households', y='population')
```

```python
# Задача 63.
# 2. Визуализировать longitude по отношению к
# median_house_value, используя линейный график

sns.relplot(x='longitude', y='median_house_value', kind='line', data=df)
```

```python
# Задача 63.
# 3. Представить гистограмму по housing_median_age

sns.histplot(data=df, x='housing_median_age')
```

```python
# Задача 63.
# 4. Изобразить гистограмму по median_house_value с
# оттенком housing_median_age

df.loc[df['housing_median_age'] <= 20, 'age_group'] = 'Молодые'
df.loc[(df['housing_median_age'] > 20) & (df['housing_median_age'] <= 50),
       'age_group'] = 'Ср. возраст'
df.loc[df['housing_median_age'] > 50, 'age_group'] = 'Пожилые'

sns.histplot(data=df, x='median_house_value', hue='age_group')
```

```python
# Задача 67.
# Создать новый столбец в таблице с пингвинами, который 
# будет отвечать за показатель длины клюва пингвина.
# high - длинный (от 42)
# middle - средний (от 35 до 42)
# low - маленький (до 35)
penguins = sns.load_dataset("penguins")
penguins.loc[penguins['bill_length_mm'] < 35, 'length_height'] = 'low'
penguins.loc[(penguins['bill_length_mm'] >= 35) & 
             (penguins['bill_length_mm'] < 42), 'length_height'] = 'middle'
penguins.loc[penguins['bill_length_mm'] >= 42, 'length_height'] = 'high'
penguins
```

```python
# Задача 69.
# 1. Изобразить гистограмму по flipper_length_mm
# с оттенком height_group. Сделать анализ
penguins = sns.load_dataset("penguins")
penguins.loc[penguins['bill_length_mm'] < 35, 'length_height'] = 'low'
penguins.loc[(penguins['bill_length_mm'] >= 35) & 
             (penguins['bill_length_mm'] < 42), 'length_height'] = 'middle'
penguins.loc[penguins['bill_length_mm'] >= 42, 'length_height'] = 'high'
sns.histplot(data=penguins, x='flipper_length_mm', hue='length_height')
```
